## 확장자명 변경

```python
import os

def change_to_jpg(folder_path):
    # 폴더 내 파일들 확인
    for filename in os.listdir(folder_path):
        # 파일의 전체 경로 생성
        file_path = os.path.join(folder_path, filename)

        # 파일인 경우에만 처리
        if os.path.isfile(file_path):
            # 파일명과 확장자 분리
            name, ext = os.path.splitext(filename)

            # 새로운 파일명 (확장자만 .jpg로 변경)
            new_filename = name + '.jpg'
            new_file_path = os.path.join(folder_path, new_filename)

            # 파일 이름 변경
            os.rename(file_path, new_file_path)
            print(f"{filename} -> {new_filename}")

# 함수 사용 예시
folder_path = '/content/train/train_defected_dataset'
change_to_jpg(folder_path)
```

```python
folder_path = '/content/validation/validation_defected_dataset'
change_to_jpg(folder_path)
```

```python
import os
import json

def update_image_extension_in_json(json_dir):

    # JSON 파일들을 불러오기
    json_files = [f for f in os.listdir(json_dir) if f.endswith(".json")]

    # 모든 JSON 파일을 순회하며 "image_name"의 확장자를 .png에서 .jpg로 변경
    for json_file in json_files:
        json_path = os.path.join(json_dir, json_file)

        # JSON 파일을 불러옴
        with open(json_path, 'r') as f:
            data = json.load(f)

        # "image_name"에서 확장자를 .jpg로 변경
        if data["image_name"].endswith(".png"):
            data["image_name"] = data["image_name"].replace(".png", ".jpg")

        # 수정된 내용을 다시 JSON 파일에 저장
        with open(json_path, 'w') as f:
            json.dump(data, f, indent=4)
```

```python
update_image_extension_in_json("/content/train/train_defected_json")
```

```python
update_image_extension_in_json("/content/validation/validation_defected_json")
```

# data.yaml 파일 생성

```python
# data.yaml 파일 생성
data_yaml = """
train: /content/train/images
val: /content/validation/images

nc: 4  # 클래스 수 (무결함, aq, st, fl)
names: [0, 1, 2, 3]  # 클래스 이름
"""

# data.yaml 파일로 저장
with open('/content/data.yaml', 'w', encoding='utf-8') as f:
    f.write(data_yaml)
```

### YOLOv8 모델 학습

```python
from ultralytics import YOLO

# YOLOv8 모델 불러오기 (Nano 모델 사용)
model = YOLO('yolov8n.pt')

# 모델 학습 시작 (에포크, 배치 사이즈, 이미지 사이즈 설정 가능)
model.train(data='/content/data.yaml', epochs=50, imgsz=640, batch=64)

# 최적 모델 저장
model.export(format='onnx')  # 필요에 따라 다른 형식으로 내보낼 수 있음
```

### YOLOv8 모델 평가

```python
# 모델 검증 수행
metrics = model.val()

# 각 평가 지표 출력
import numpy as np

# Precision, Recall, mAP 계산
precision = np.mean(metrics.box.p)
recall = np.mean(metrics.box.r)
map50 = np.mean(metrics.box.map50)
map50_95 = np.mean(metrics.box.map)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"mAP@0.5: {map50:.4f}")
print(f"mAP@0.5:0.95: {map50_95:.4f}")
```
